```python
df['activeDate_dt'] = pd.to_datetime(df['activeDate'], unit='ms')

def create_graph_date(date_input):
    """
    Format date_input: YYYY-MM-DD
    """
    if isinstance(date_input, str):
        ls_check_date = date_input.split('-')
        if len(ls_check_date) == 3:
            year, month, day = ls_check_date
            if (len(year) == 4 and len(month) == 2 and len(day) == 2 and
                1 <= int(month) <= 12 and 1 <= int(day) <= 31):
            
                if (int(month) in [4, 6, 9, 11] and int(day) > 30) or (int(month) == 2 and int(day) > 29):
                    print("ÿ™ÿßÿ±€åÿÆ ŸÜÿßÿØÿ±ÿ≥ÿ™ ÿßÿ≥ÿ™.")
                    return
                
                date_obj = dt.datetime.strptime(date_input, '%Y-%m-%d')
                filtered = df[df['activeDate_dt'].dt.date == date_obj.date()]

                if filtered.empty:
                    print("ÿ™ÿßÿ±€åÿÆ ŸÖŸàÿ¨ŸàÿØ ŸÜ€åÿ≥ÿ™")
                else:
                    activation_counts_sort = filtered['activeDate_dt'].dt.date.value_counts()

                    plt.figure(figsize=(5,6))

                    title = get_display(arabic_reshaper.reshape('ÿ™ÿπÿØÿßÿØ ŸÖÿ¥ÿ™ÿ±⁄©€åŸÜ ÿØÿ± ÿ™ÿßÿ±€åÿÆ ŸÖŸàÿ±ÿØŸÜÿ∏ÿ±'))
                    plt.title(title)
                    xlabel = get_display(arabic_reshaper.reshape('ÿ™ÿßÿ±€åÿÆ ŸÅÿπÿßŸÑ‚Äåÿ≥ÿßÿ≤€å'))
                    plt.xlabel(xlabel)
                    ylabel = get_display(arabic_reshaper.reshape('ÿ™ÿπÿØÿßÿØ ŸÖÿ¥ÿ™ÿ±⁄©€åŸÜ'))
                    plt.ylabel(ylabel)

                    x = [i.strftime('%Y-%m-%d') for i in activation_counts_sort.index]
                    y = activation_counts_sort.values

                    bars = plt.bar(x, y, color='skyblue')

                    for bar in bars:
                        yval = bar.get_height()
                        plt.text(bar.get_x() + bar.get_width() / 2, yval, int(yval), ha='center', va='bottom')

                    
                    plt.tight_layout()
                    plt.show()
            else:
                print("ŸÅÿ±ŸÖÿ™ ÿ™ÿßÿ±€åÿÆ ŸÜÿßÿØÿ±ÿ≥ÿ™ ÿßÿ≥ÿ™.")
        else:
            print("ŸÅÿ±ŸÖÿ™ ÿ™ÿßÿ±€åÿÆ ŸÜÿßÿØÿ±ÿ≥ÿ™ ÿßÿ≥ÿ™.")
    else:
        print("ŸÅÿ±ŸÖÿ™ ÿ™ÿßÿ±€åÿÆ ŸÜÿßÿØÿ±ÿ≥ÿ™ ÿßÿ≥ÿ™.")


create_graph_date('2007-05-19')
```

```python
customer_ID = df['customerID'].values
service_Number = df['serviceNumber'].astype(str).values
rond=[]
df_round = pd.DataFrame({
    'customerID': customer_ID,
    'serviceNumber': service_Number,
    'mirror': ['-' for _ in range(len(service_Number))],
    'birthday_date': ['-' for _ in range(len(service_Number))],
    'three_number_seqential': ['-' for _ in range(len(service_Number))],
    'pre_number': ['-' for _ in range(len(service_Number))],
    'last_four_number_less_different': ['-' for _ in range(len(service_Number))]
})

mirror_mask = np.zeros(len(service_Number), dtype=bool)
birthday_mask = np.zeros(len(service_Number), dtype=bool)
three_seq_mask = np.zeros(len(service_Number), dtype=bool)
pre_number_mask = np.zeros(len(service_Number), dtype=bool)
last_four_mask = np.zeros(len(service_Number), dtype=bool)

for j, i in enumerate(service_Number):
    # 1. ÿπÿØÿØ ŸÖŸÇŸÑŸàÿ® €åÿß ÿß€å€åŸÜŸá ÿß€å ÿØÿ± Ÿàÿ≥ÿ∑ €åÿß ÿßÿÆÿ± ÿ¥ŸÖÿßÿ±Ÿá
    if (i[3:5] == i[6:4:-1]) or (i[6:8] == i[9:7:-1]) or (i[0:2] == i[3:1:-1]) or (i[1:3] == i[4:2:-1]) or (i[2:4] == i[5:3:-1]) or (i[4:6] == i[7:5:-1]) or (i[5:7] == i[8:6:-1]):
        mirror_mask[j] = True
        if i not in rond:
            rond.append(i)

    # 2. ÿ™ÿßÿ±€åÿÆ ÿ™ŸàŸÑÿØ ÿ¥ŸÖÿ≥€å €åÿß ŸÖ€åŸÑÿßÿØ€å ÿØÿ± ÿßÿÆÿ± €åÿß ÿßŸàŸÑ ÿ¥ŸÖÿßÿ±Ÿá
    if (i[6:8] == '19') or (i[6:8] == '20' and int(i[8:10]) < 27) or (i[6:8] == '13') or (i[6:8] == '14' and int(i[8:10]) < 6) or (i[2:4] == '19') or (i[2:4] == '20' and int(i[4:6]) < 27) or (i[2:4] == '13') or (i[2:4] == '14' and int(i[4:6]) < 6):
        birthday_mask[j] = True
        if i not in rond:
            rond.append(i)

    # 3. ÿ≥Ÿá ÿπÿØÿØ ŸÖÿ™ŸàÿßŸÑ€å ÿØÿ± Ÿàÿ≥ÿ∑ ÿ¥ŸÖÿßÿ±Ÿá ŸÖÿ´ŸÑ 123
    if (abs(int(i[4]) - int(i[3])) == 1 and abs(int(i[5]) - int(i[4])) == 1 and abs(int(i[5]) - int(i[3])) == 2):
        three_seq_mask[j] = True
        if i not in rond:
            rond.append(i)

    # 4. ÿ™⁄©ÿ±ÿßÿ± Ÿæ€åÿ¥ ÿ¥ŸÖÿßÿ±Ÿá ÿßŸàŸÑ €åÿß ÿßÿÆÿ± ÿ¥ŸÖÿßÿ±Ÿá ŸÖÿ´ŸÑ 914 917
    if i[3] == '9' or i[6:8] == '09':
        pre_number_mask[j] = True
        if i not in rond:
            rond.append(i)

    # 5. ⁄ÜŸáÿßÿ±ÿ¥ŸÖÿßÿ±Ÿá ÿßÿÆÿ± €å⁄©€å ÿßÿÆÿ™ŸÑÿßŸÅ ÿØÿßÿ±ŸÜÿØ
    if (abs(int(i[6:8]) - int(i[8:10])) == 1) or (i[6] == i[8]) or ((int(i[6:8]) - int(i[8:10])) % 10 == 0):
        last_four_mask[j] = True
        if i not in rond:
            rond.append(i)


df_round.loc[mirror_mask, 'mirror'] = '*'
df_round.loc[birthday_mask, 'birthday_date'] = '*'
df_round.loc[three_seq_mask, 'three_number_seqential'] = '*'
df_round.loc[pre_number_mask, 'pre_number'] = '*'
df_round.loc[last_four_mask, 'last_four_number_less_different'] = '*'
```
