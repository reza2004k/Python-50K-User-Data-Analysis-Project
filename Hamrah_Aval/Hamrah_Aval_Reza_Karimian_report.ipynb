گزارش پروژه تحلیل اطلاعات مشترکین همراه اول

معرفی پروژه : 

در این پروژه، اطلاعات 50 هزار مشترک همراه اول مورد تحلیل قرار گرفته است. داده‌ها از یک فایل CSV با نام 50K_sample.csv خوانده شده‌اند. هدف اصلی، بررسی آماری ویژگی‌های مشترکین و استخراج بینش‌هایی در مورد توزیع جنسیت، شهر محل سکونت، نوع سیم‌کارت، تاریخ فعال‌سازی، نام‌ها و شناسایی شماره‌های خاص و رند بوده است.

مراحل تحلیل و نتایج

1. خواندن و بررسی اولیه داده‌ها
با استفاده از کتابخانه pandas، داده‌ها بارگذاری شدند.

ساختار داده‌ها شامل ستون‌هایی مانند: serviceNumber (شماره موبایل)، firstName (نام)، addr2 (نام شهر)، gender (جنسیت)، paymentType_value (نوع سیم‌کارت)، activeDate (تاریخ فعال‌سازی) است.

2. تحلیل اکتشافی داده ها به صورت بررسی داده های گشده و داده های پرت و encoding کردن برخی از داده های مورد نیاز صورت گرفته است .

در مجموع در 6 تا از ستون ها داده های گمشده وجود دارد.
که درصد برخی از انها کمتر از یک و برخی دیگر بالای 80 است .

در قسمت داده های پرت برای ستون شهر ها داده هایی وجود داشتند که از الگو پیروی نکرده یا به صورت حروف نبودند و یا به صورت حروف فارسی نبودند .

در قسمت encoding کردن داده ها از دسته یندی نرمال استفاده شده است .

3. توزیع مشترکین به تفکیک شهر

دو نمودار ستونی افقی و عمودی رسم شده که نشان می‌دهد بیشترین تعداد مشترکین مربوط به شهرهایی مانند تهران، البرز و قم است.

نام های فارسی به کمک arabic_reshaper و bidi به صورت صحیح فارسی نمایش داده شده‌اند.

4. تحلیل نوع سیم‌کارت

توزیع سیم‌کارت‌ها (دائمی و اعتباری) مورد بررسی قرار گرفته و با نمودار ستونی تعداد مشترکین هر کدام قابل مشاهده است .

5. نام‌های پر تکرار

با استفاده از تابع value_counts() نام‌ها شناسایی شدند.

نمودار میله‌ای نمایش دهنده پرتکرارترین نام‌ها در میان مشترکین رسم شده است.
و همچنین نموداری که میتوان فقط در آن تعداد نام مورد نظر خود را مشاهده کرد .

6. تحلیل جنسیتی

درصد جنسیت مرد و زن با استفاده از نمودار میله ای نمایش داده شده است .
توزیع بین جنسیت‌ها نسبتا متوازن بوده، اما با تفاوت اندکی مرد ها بیشتر هستند .

7. بررسی تاریخ‌های فعال‌سازی

تاریخ فعال‌سازی به فرمت مناسب تاریخ تبدیل شده است . 
با استفاده از نمودار میله ای، روند تغییرات تعداد فعال‌سازی‌ها در طول زمان نشان داده شده است.
و همچنین میتوان در نموادری فقط تعداد مشترکین روزهایی که مد نظر هستن را مشاهده کرد.

8. شناسایی شماره‌های رند

الگوریتم هایی برای بررسی و استخراج شماره‌های خاص (رند یا با الگوهای تکراری) طراحی شده است.
و با سرچ شماره مدنظر میتوان میزان رند بودن و نوع رند بودن را مشاهده کرد .

9. بررسی تاریخ تولد مشترکین

با نمودار میله ای میتوان تعداد متولدین مشترکین را در یک روز خاص بررسی کرد .

10. وضعیت سیمکارت ها 

با استفاده از تابع value_counts() توانستیم به فعال بودن همه سیمکارت ها پی ببریم .

11. بررسی وجود تاریخ تولد مشترکین در شماره هایشان

با بررسی های متعددی میتوان در جداول رسم شده افرادی که سال و یا سال و ماه و یا ماه و روز و یا سال و ماه و روز تولدشان در شماره هایشان است را مشاهده کرد .

13. بررسی سیمکارت های متعلق به یک فرد

با بررسی ستون های کد ملی و شماره تماس ها متوجه میشویم که هر فرد تنها یک سیمکارت دارد .

این توابع امکان بررسی دقیق‌تر و تعامل با داده را فراهم می‌کنند.

جمع‌بندی : 

پروژه حاضر تحلیل جامعی از ویژگی‌های مشترکین همراه اول را ارائه می‌دهد. استفاده از ابزارهای قدرتمند پایتون مانند pandas و matplotlib همراه با نمایش فارسی صحیح داده‌ها، باعث شده این پروژه از لحاظ تحلیلی و گرافیکی بسیار کارآمد باشد.
